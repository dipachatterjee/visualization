```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)

# scattergraph
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)

# histograph/bar graph
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

# line graph
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y, edgecolor='black')

# histogram/bar plot
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
fig, ax = plt.subplots(figsize=(5, 3)) # setting figure size
ax.plot(x,y) # plot it
ax.set_title('Total growth over time', fontdict = font1) # setting title with font1 specifications
ax.set_ylabel('Total growth', fontdict = font2) # setting y-label with font2 specifications
ax.set_xlabel('Years since start', fontdict = font2) # setting x-label with font2 specifications
fig.tight_layout()
```

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
  
sns.set_theme(style="darkgrid")
df = sns.load_dataset('iris')
 
# Make default density plot
sns.kdeplot(df['sepal_width'])
plt.show()
```

```python
df = sns.load_dataset('iris')

sns.set_theme(style="white") #set theme to white, with no gridline

sns.kdeplot(
    df['sepal_width'],
    fill=True,
    color="green",
    alpha=0.5,
    linewidth=0
) # graph for the fill, no outline

sns.kdeplot(
    df['sepal_width'],
    fill=False,
    color="black",
    linewidth=2
) # graph for the line only

# set font specifications
font1 = {'family':'sans-serif','color':'blue', 'weight':'bold', 'size':20}
font2 = {'family':'sans-serif','color':'black','weight':'bold', 'size':14}

# set titles and labels with specified fonts
plt.title("Trial Histograph Graph", fontdict=font1)
plt.xlabel("Trial X Axis", fontdict=font2)
plt.ylabel("Trial Y Axis", fontdict=font2)

plt.show()
```
