```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50) # random array from 0-49 for x-axis
y1 = np.random.randint(0, 100,50) # first data set (0 to 100, 50 numbers)
y2 = np.random.randint(0, 100,50) # second set of data
```

```python
#make 2 line plots in one graph

fig, ax = plt.subplots(figsize=(5,3)) #figure object and axis object using subplot; and set size...5 is width and 3 is height
ax.plot(x,y1) #plot first dataset
ax.plot(x,y2) #plot second dataset
#fig.show()
```

```python
#add legend using ax.plot and ax.legend
#to set the axis labels use ax.set_xlabel/y1label/y2label

fig, ax = plt.subplots (figsize=(5,3))
ax.plot(x,y1, label = 'Person 1')
ax.plot(x,y2, label = 'Person 2')
ax.legend(loc='best') #set legend at what you specify...loc = location

#"best"
#"upper right"
#"upper left"
#"lower left"
#"lower right"
#"right"
#"center left"
#"center right"
#"lower center"
#"upper center"
#"center"
```

```python
# modify legend

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
    frameon = True, #add/remove frame around the legend
    fontsize = 12, #change font size
    ncol = 2, #specify number of columns
    shadow = True ) #shadows in legend box
```

```python
# move legend outside plot area using bbox_to_anchor

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left', #attaching legend's upper left corner to the plot's top right corner
    bbox_to_anchor =(1, 1)) # (1,1) = top right

#(0,0) usually origin (bottom left) or (1,1) top right...usually use these only
```

```python
# annotating

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text (10, 95, 'This value is important!', #x=10, y=95
         ha='center', #horizontal alignment of text of the sentence to the specified x,y location
         color = 'red', #modify font colour
         size = 20) #modify font size
```

```python
fig,ax=plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label='Person 1')
ax.scatter(x,y2,label="Person 2")
ax.legend(loc='lower right',prop={"family":"serif","size":20}) # prop to change the characteristics of the legend
```

```python
# positioning annotation

fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData) # to data
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes) # to axis...0.5 = half way (50%) between the x axis, 0.1 = 10% of the y axis
ax.text(0.5, 0.1, ". Figure:(0.5, 0.1)", transform=fig.transFigure) # to the whole figure (whole window)
```

```python
# annotate with arrows using arrowprops = a dictionary with many different arguments
# use ax.annotate if you want to specify a data point, otherwise can use ax.text

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', 
            xy=(10, 95), #data point we want to focus on
            xytext=(25, 94), #where the text will sit
            arrowprops=dict(facecolor='black')) #facecolor = arrowhead colour
```

```python
# modify arrow

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!',
            xy=(10, 95), xytext=(20, 94),
            arrowprops = dict(arrowstyle = "<->",  # <-, ->, <->, wedge
                              color = "hotpink"))
```

```python
# removing tick marks and labels with set_major_locator (ticks) and set_major_formatter (labels)


fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = 'Person 1')
ax.scatter(x,y2, label = 'Person 2')
ax.legend(loc='lower right')

#ax.yaxis.set_major_locator(plt.NullLocator()) # ticks & labels dissapear
#ax.xaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter()) # labels dissapear
```

```python
# limiting tick marks using MaxNLocator

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(10)) # max number of ticks on x-axis is whatever # you specify
```

```python
# tick mark intervals using MultipleLocator

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(10))
```

```python
# rotating x-axis labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')
#plt.xticks(rotation=-45, ha='left')
```
